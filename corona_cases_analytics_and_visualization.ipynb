```python
#caluculating all 'name' column total death and cases...


all_province_cases = {}
all_province_death = {}
for Name in df['Name']:
    if Name not in all_province_cases and Name != 'Americas 	':
        all_province_cases[Name] = df.loc[df['Name']==Name,'Cases - newly reported in last 24 hours'].sum()
        all_province_death[Name] = df.loc[df['Name']==Name,'Deaths - newly reported in last 7 days per 100000 population'].sum()

all_province_death
```

```python
all_province_cases
```

```python
plt.scatter(all_province_death.values(),all_province_death.values())
plt.xlabel('Deaths a/c province')
plt.ylabel('cases a/c province')
plt.show()
```
